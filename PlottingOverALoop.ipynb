```python
import numpy as np

import matplotlib.pyplot as plt
```

Let's say I want to know the maximum value of the function:

$$
f(x) = \exp(-x) \sin(k x)
$$

for different values of $k$, over the range $x \in [0, 5]$. 

First we'll define a function $f(x, k)$:

```python
def f(x, k):
    return np.exp(-x)*np.sin(k*x)
```

Now let's write a function which calculates what we want:

```python
#In principle, we could use some kind of minimizing/optimizing routine from scipy here, 
#to find the maximum, but we'll do it roughly (i.e. calculating over a grid of x-values)
def calc_max(k):
    x_list = np.linspace(0, 5, 1000)
    f_list = f(x_list, k)
    return np.max(f_list)
```

Now we loop over different $k$-values and calculate want we want:

```python
k_list = np.arange(10)
max_list = np.zeros(len(k_list))

for i, k in enumerate(k_list):
    max_list[i] = calc_max(k)
```

And now we plot!

```python
plt.figure()

plt.plot(k_list, max_list)

plt.xlabel(r"$k$")
plt.ylabel(r"$\mathrm{max} f$")

plt.show()
```

Note that we didn't need to actually plot the functions to calculate the maximum, but let's do it separately anyway, just to check that everything makes sense:

```python
plt.figure(figsize=(7,5))

x_list = np.linspace(0, 5, 1000)
for i, k in enumerate(k_list):
    f_list = f(x_list, k)
    plt.plot(x_list, f_list, label=r'$k = ' + str(int(k)) + '$')
    
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x, k)$")
    
plt.legend(loc='upper right')
    
plt.show()
```

Yep! As we increase $k$, the maximum value of the function $f(x)$ increases, so it looks like everything makes sense!
